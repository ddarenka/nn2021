### Lab #2:

```python
from sklearn import datasets
```

```python
#student index number = 10
n_rows = 100000
n_cols = 30-10
```

### make_regression():

```python
data = datasets.make_regression(
    n_samples=n_rows, #the number of samples (rows)
    n_features=n_cols, #the number of features (X-columns)
    n_informative=int(n_cols/2), # the number of features used to build the linear model (to generate the output)
    n_targets = int(n_cols/n_cols), #the dimension of the y output vector
    bias=150,
    effective_rank = None, #the input set is well conditioned, centered and gaussian with unit variance
    tail_strength = 0.7, #the relative importance of the fat noisy tail of the singular values profile
    noise = 0.9, #the standard deviation of the gaussian noise applied to the output
    shuffle = True,
    coef = False,
    random_state=42)
```

### Data:

```python
import pandas as pd
df = pd.DataFrame(data[0])
df = df.add_prefix('x') #adding prefix to column names
df['y'] = data[1]
df.head()
```

```python
df.to_csv('data/data.csv', sep=';', index=False)
```
