```python
import pandas as pd
```

### Data import:

```python
df = pd.read_csv('data/data.csv', sep=';')
df.head()
```

### X-Y split:

```python
X_df = df.iloc[:,:-1]
y_df = df.iloc[:,-1]
```

### Train-test split:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, train_size=0.67, random_state=42)
```

### Model:

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

### Vizualization:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
figure(figsize=(20, 10), dpi=80)

X_graph = np.arange(len(y_test))

plt.scatter(X_graph, y_test, color = 'olive')
plt.scatter(X_graph, y_pred, color = 'palevioletred')
ax = plt.axes()
ax.set_facecolor("white")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

### MSE:

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

> ### Conclusion:
_The data was generated using the Sklearn 'make_regression' function with as many adjusted parameters as possible and then exported to CSV format. This data, split into the train-test X-y datasets, then was used to build the linear regression model. The results were visualized with the matplotlib and, according to the graph and mean squared error equal to 0.81, we can conclude that the model prediction accuracy is quite high._
